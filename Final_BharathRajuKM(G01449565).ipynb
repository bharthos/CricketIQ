```python
import pandas as pd
import matplotlib.pyplot as plt
import os
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import LabelEncoder


os.getcwd()
os.chdir('D:/AIT580/Final Project')

df = pd.read_csv('MenT20ITeamMatchResults_Cleaned.csv')

df.head(10)
df.columns

df.info()

df['result_code'] = pd.Categorical(df['Result']).codes
result_code_dict = dict(zip(df['Result'], pd.Categorical(df['Result']).codes))
result_code_dict

# Feature Engineering

df['result_cat_code'] = pd.Categorical(df['Result_cat']).codes
result_cat_code_dict = dict(zip(df['Result_cat'], pd.Categorical(df['Result_cat']).codes))
result_cat_code_dict
df['month_code'] = pd.Categorical(df['Match Month']).codes
month_code_dict = dict(zip(df['Match Month'], pd.Categorical(df['Match Month']).codes))
month_code_dict

df['ground_code'] = pd.Categorical(df['Ground']).codes
ground_code_dict = dict(zip(df['Ground'], pd.Categorical(df['Ground']).codes))
ground_code_dict


df['home_away_code'] = pd.Categorical(df['Home/Away']).codes
home_away_code_dict = dict(zip(df['Home/Away'], pd.Categorical(df['Home/Away']).codes))
home_away_code_dict

df['country_code'] = pd.Categorical(df['Country']).codes
country_code_dict = dict(zip(df['Country'], pd.Categorical(df['Country']).codes))
country_code_dict

# Predictors
features = df[['result_cat_code', 'month_code',
'ground_code', 'home_away_code', 'country_code']]

# Response variable
labels = df[['result_code']]

labels = labels.values.ravel()

# test-train split for cross validation
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=42)

# Random Forest
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)

# Train the classifier
rf_classifier.fit(X_train, y_train)

# Make predictions on the test set
y_pred = rf_classifier.predict(X_test)


# Evaluate the performance
accuracy = accuracy_score(y_test, y_pred)
accuracy
```

```python
feature_importance = rf_classifier.feature_importances_

# Create a DataFrame to store feature names and their importance scores
feature_importance_df = pd.DataFrame({'Feature': features.columns, 'Importance': feature_importance})

# Sort the DataFrame by importance in descending order
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)

# Plot the feature importance graph
plt.figure(figsize=(10, 6))
plt.bar(feature_importance_df['Feature'], feature_importance_df['Importance'])
plt.xlabel('Feature')
plt.ylabel('Importance')
plt.title('Random Forest Feature Importance')
plt.show()
```
